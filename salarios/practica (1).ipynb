#### Identifica los outliers de nuestro dataset de forma gráfica: 

```python
import math
def identificar_outliers(dataframe):
    """
    Esta función crea un gráfico de caja (boxplot) para cada columna numérica 
    en un DataFrame para ayudar a identificar los outliers.
    """
    # Seleccionamos solo las columnas numéricas
    numeric_cols = dataframe.select_dtypes(include=np.number).columns

    # Creamos los subplots
    ncols = 4
    nrows = math.ceil(len(numeric_cols) / ncols) 
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 5*nrows))

    # Si el dataframe tiene menos de 4 columnas numéricas, axes no será una lista de listas, por lo que se convierte en una para poder iterar
    if nrows == 1 or ncols == 1:
        axes = [axes]

    for i, col in enumerate(numeric_cols):
        row = i // ncols
        col_idx = i % ncols
        # Comprueba si la columna tiene más de un valor único para evitar errores
        if len(dataframe[col].unique()) > 1:
            sns.boxplot(x=dataframe[col], data=dataframe, ax=axes[row][col_idx])
            axes[row][col_idx].set_title(f"Gráfico: {col}", color="DarkBlue")
        else:
            print(f"La columna {col} tiene un solo valor único y no se puede graficar")

    plt.tight_layout()
    plt.show()

identificar_outliers(df)
```

#### Preguntas a resolver usando las librerías de creaciones gráficas vistas en clase: 

¿Cuál es el salario promedio por nivel de experiencia?

```python
#sacar de experience_level el salary_in_usd medio

fig = plt.figure(figsize=(18, 6))
sns.boxplot(x="experience_level", y="salary_in_usd", data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
salario_medio = df.groupby('experience_level')['salary_in_usd'].mean()
fig = plt.figure(figsize=(18, 6))
plt.xticks(rotation=90)
salario_medio.plot.bar()
```

¿Cómo ha cambiado el salario medio con el paso del tiempo (work_year)?

```python
plt.figure(figsize=(12, 4))
salary_order = sns.lineplot(x=df["work_year"] , y= df["salary_in_usd"], color="red") 
salary_order.set_xlabel("años")
salary_order.set_ylabel("salario")
plt.show()
```

¿Cuál es la proporción de trabajo remoto entre diferentes roles de trabajo?

```python
average_remote_ratio_per_job = df.groupby(["job_title"])["remote_ratio"].mean().reset_index()
plt.figure(figsize=(18, 35))
sns.barplot(data = average_remote_ratio_per_job , x = 'remote_ratio', y= 'job_title', palette="rocket")
plt.title("Evolution of the remote ratio in the different Job titles")
plt.ylabel("Remote Ratio (%)")
plt.xlabel("job_titles")
plt.show()
```

¿Existe una relación entre el tamaño de la empresa y el salario?

```python
df["company_size_num"] = df["company_size"].map({"S":1, "M":2, "L":3 } )
corr = df[["salary_in_usd", "company_size_num"]].corr()
print(corr)
```

```python
sns.heatmap(corr, annot=True, linewidths=.5, square=True, cmap='coolwarm')
```

¿Cuál es el país con más trabajadores remotos?

```python
remote_work_by_country = df.groupby("employee_residence")['remote_ratio'].sum().reset_index()
plt.figure(figsize=(15, 8))
sns.barplot(data=remote_work_by_country, x="employee_residence", y="remote_ratio")
plt.xticks(rotation=90)
plt.show()
```

¿Cómo ha cambiado la proporción de trabajo remoto con el paso del tiempo?

```python
plt.figure(figsize=(6, 3))
sns.lineplot(x= "work_year", y="remote_ratio", data=df , color= "green")
plt.ylabel("Propoción de trabajadores remoto")
plt.xlabel("paso del tiempo")
plt.show()
```

¿Cuál es el país con los salarios promedio más altos en USD?

```python
avg_salary_by_country = df.groupby('employee_residence')['salary_in_usd'].mean().reset_index()

plt.figure(figsize=(16,8))
order = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).index.values
sns.barplot(data = df,
            x='employee_residence',
            y='salary_in_usd',
            order=order,
              ).set_title('Países más salarios más altos en USD por país')
plt.xticks(rotation=90)
plt.ylabel('Salario')
plt.xlabel('País')
plt.show()
```

¿Cómo varía el salario promedio entre los diferentes tipos de empleo, fulltime, partime ect?

```python
avg_salary_by_employment = df.groupby('employment_type')['salary_in_usd'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_salary_by_employment.sort_values('salary_in_usd', ascending=False), 
            x='employment_type', y='salary_in_usd', palette='cool', edgecolor='black')
plt.title('Salario medio por tipo de empleo')
plt.xlabel('Tipo de empleo')
plt.ylabel('Salario medio (USD)')
```

#### Crear un top10 con los empleos mas demandados:

```python
top10_empleos = df.job_title.value_counts()[:10].index.tolist()
plt.figure(figsize=(8, 8))
sns.countplot(data=df[df.job_title.isin(top10_empleos)], y='job_title', palette='Set3')
plt.title('Top 10 de empleos en Data')
plt.xlabel('Cantidad de empleados')
plt.ylabel('Título del empleo')
plt.show()
```

####  Crear un top10 con los empleos mas cotizados:

```python
top10_cotizados = df.groupby('job_title')['salary_in_usd'].mean().reset_index()
top10_cotizados = top10_cotizados.sort_values(by='salary_in_usd', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top10_cotizados, y='job_title', x='salary_in_usd', palette='Set3')
plt.title('Top 10 empleos más cotizados')
plt.xlabel('Salario medio (USD)')
plt.ylabel('Título del empleo')
plt.show()
```
